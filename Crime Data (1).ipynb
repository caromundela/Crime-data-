```python
import pandas as pd
```

```python
import numpy as np
```

```python
df= pd.read_csv('https://raw.githubusercontent.com/caromundela/Crime-data-/main/NYPD_Arrest_Data__Year_to_Date_.csv')
```

```python
#Analyze crime data 
df.head(10)
```

```python
#display number of rows and columns in the data set 
df.shape
```

```python
#rename columns 
df= df.rename(columns={'OFNS_DESC': 'Charge', 'ARREST_BORO': 'Borough', 'PERP_SEX': 'Gender','PERP_RACE': 'Race'})
df.tail(10)
```

```python
#drop select columns 
drop_cols =['ARREST_KEY', 'PD_CD', 'KY_CD', 'LAW_CAT_CD', 'JURISDICTION_CODE', 'X_COORD_CD', 'Y_COORD_CD', 'Latitude', 'Longitude', 'New Georeferenced Column']
```

```python
df.drop(drop_cols, inplace =True, axis=1)
df.head(10)
```

```python
df.groupby(['Charge'])['Race'].count()
```

```python
#clean up the null 
df.isnull().sum()
```

```python
#filter by borough Kings County 
df[df['Borough'] =='K']
```

```python
#Analyze number of burglaries in Kings County 
df.loc[(df['Borough'] == 'K') & (df['Charge'] == 'BURGLARY')]
```

```python
#create new DF for burglaries in Kings County 
df1= df.loc[(df['Borough'] == 'K') & (df['Charge'] == 'BURGLARY')]
df1.head()
```

```python
#Kings County burglaries by age group 
df1.groupby(['AGE_GROUP'])['Charge'].count()
```

```python
#Kings County burglaries by race 
df1.groupby(['Race'])['Charge'].count()
```

```python
#Kings County burglaries by gender 
df1.groupby(['Gender'])['Charge'].count()
```

```python
#Insert Visual 
import matplotlib.pyplot as plt 
import numpy as np
```

```python
#Graph men and women in Kings County that committed burglaries by race 
x = np.array(["Asian/Pacific Islander", "Black", "Black Hispanic", "Unknown", "White", "White Hispanic"])
y= np.array([14, 264, 39, 4, 85, 135])

plt.barh(x, y, color = "green")
plt.show()
```

```python
#Analyze hoow many men committed burglaries in each borough 
df2= df.loc[(df['Gender'] == 'M') & (df['Charge'] == 'BURGLARY')]
df2.head(10)
```

```python
df2.shape
```

```python
#View by race 
df2.groupby(['Race'])['Charge'].count()
```

```python
#vuew by borough 
df2.groupby(['Borough'])['Charge'].count()
```

```python
#Graph men charged with burglary in all boroughs by race 
x = np.array(["Asian/Pacific Islander", "Black", "Black Hispanic", "Unknown", "White", "White Hispanic"])
y= np.array([49, 712, 132, 8, 159, 448])

plt.barh(x, y, color = 'yellow')
plt.title("Males who Committed Burglaries in NYC by Race", fontsize = 18)
plt.show()
```

```python
#Males who committed the most burglaries by county 
x = np.array(["Bronx", "Kings", "Manhattan", "Queens", "Staten Island"])
y= np.array([276, 477, 516, 204, 35])

plt.bar(x, y, width = 0.3, color = 'orange')
plt.title("Males who Committed Burglaries By County", fontsize = 18)
plt.show()
```

```python
#display subplot 

#plot 1 Kings County:
x = np.array(["Asian/Pacific Islander", "Black", "Black Hispanic", "Unknown", "White", "White Hispanic"])
y= np.array([7, 124, 12, 3, 36, 42])

plt.subplot(1, 6, 1)
plt.plot(x,y)
plt.title('Kings')
plt.xlabel('Race', fontsize = 12)

plt.xticks(rotation = 90)



#plot 2 Manhattan County: 
x = np.array(["Asian/Pacific Islander", "Black", "Black Hispanic", "Unknown", "White", "White Hispanic"])
y= np.array([7, 153, 19, 2, 21, 95])

plt.subplot(1, 6, 2)
plt.plot(x,y)
plt.title('Manhattan')
plt.xlabel('Race', fontsize = 12)

plt.xticks(rotation = 90)


#plot 3 Bronx County
x = np.array(["Asian/Pacific Islander", "Black", "Black Hispanic", "Unknown", "White", "White Hispanic"])
y= np.array([3, 57, 37, 1, 1, 50])

plt.subplot(1, 6, 3)
plt.plot(x,y)
plt.title('Bronx')
plt.xlabel('Race', fontsize = 12)

plt.xticks(rotation = 90)

plt.subplots_adjust(left=0.5,
                    bottom=0.5,
                    right=1.7,
                    top=1.2,
                    wspace=0.6,
                    hspace=1.0)
plt.show()
```
